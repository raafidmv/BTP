```python
# Define decision variables
transport_vars = LpVariable.dicts("Transport", [(i, j, k, t, w) for i in distribution_centers 
                                                  for j in ['plant'] 
                                                  for k in parts 
                                                  for t in companies
                                                  for w in weeks], lowBound=0, cat='Integer')

# Define the number of parts stored in each distribution center for each part type and each week
# inventory_vars = pulp.LpVariable.dicts("Inventory", ((i, j, k) for i in distribution_centers for j in parts for k in weeks), lowBound=0, cat='Continuous')

```

```python
# Define objective function
prob += lpSum(transport_vars[(i, 'plant', k, t, w)] * cost_price[(k, t)]  for i in distribution_centers 
                                                                 for k in parts 
                                                                 for t in companies
                                                                 for w in weeks)
```

```python
# Define demand constraints
for i in distribution_centers:
    for k in range(len(parts)):
        sum_d=0
        sum_t=0
        for w in range(len(weeks)):
            prob += lpSum(transport_vars[(i, 'plant', parts[k], t,weeks[w])] for t in companies) >= demand[('center_'+str(i), parts[k],weeks[w])]-(sum_t-sum_d)
            sum_d=sum_d+demand[('center_'+str(i), parts[k], weeks[w])]
            sum_t=sum_t+sum([transport_vars[(i, 'plant', parts[k], t, weeks[w])] for t in companies])
```

```python
# Define transportation capacity constraints
for t in companies:
    prob += lpSum(transport_vars[(i, 'plant', k, t, w)] * weight_per_unit[k] for i in distribution_centers 
                                                                               for k in parts
                                                                               for w in weeks) <= transport_capacity[t] * truck_capacity
```

```python
# Define production capacity constraints
for k in parts:
    prob += lpSum(transport_vars[(i, 'plant', k, t, w)] for i in distribution_centers 
                                                        for t in companies
                                                        for w in weeks) <= production_capacity[k]
```

```python
# Solve the problem
prob.solve()

# Print the optimal solution
print("Total cost:", value(prob.objective))
for v in prob.variables():
    if v.varValue > 0:
        print(v.name, "=", v.varValue)
```
